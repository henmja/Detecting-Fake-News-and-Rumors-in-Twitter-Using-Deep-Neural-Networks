# Preprocessing

## Read Data:

```python
import os, json
from os import path
import pandas as pd
import numpy as np
import glob
pd.set_option('display.max_columns', None)

fakeArticlesPD = pd.DataFrame()
realArticlesPD = pd.DataFrame()

for i in range(2):
    if i == 0:
        authenticity = 'fake'
    else:
        authenticity = 'real'
    rootdir = 'C:/Users/henri/FakenewsData/fakenewsnet_data/fakenewsnet_data/politifact/'+authenticity
    
    for subdir, dirs, articles in os.walk(rootdir):
        for article in dirs:
            temp = pd.DataFrame()
            path_to_json = 'C:/Users/henri/FakenewsData/fakenewsnet_data/fakenewsnet_data/politifact/' + authenticity + '/'+ article + '/tweets' 
            json_pattern = os.path.join(path_to_json,'*.json')
            file_list = glob.glob(json_pattern)

            for file in file_list:
                try:
                    data = pd.read_json(open(file, "r", encoding="utf8", errors="surrogateescape"), lines=True)
                    if authenticity == 'fake':
                        fakeArticlesPD = fakeArticlesPD.append(data, ignore_index = True)
                    else:
                        realArticlesPD = realArticlesPD.append(data, ignore_index = True)
                    break
                except ValueError:
                    print("ValueError")
        #print(temp["user"][0]['id'])
        #print(temp['retweets'][13][0]['user']['id'])
     
            path_to_json = 'C:/Users/henri/FakenewsData/fakenewsnet_data/fakenewsnet_data/politifact/'+authenticity+'/'+article+'/retweets' 
            json_pattern = os.path.join(path_to_json,'*.json')
            file_list = glob.glob(json_pattern)

            for file in file_list:
            #print(file)
                try:
                    data = pd.read_json(open(file, "r", encoding="utf8", errors="surrogateescape"), lines=True)
                    if authenticity == 'fake':
                        fakeArticlesPD = fakeArticlesPD.append(data, ignore_index = True)
                    else:
                        realArticlesPD = realArticlesPD.append(data, ignore_index = True)
                    break
                except ValueError:
                    print("ValueError")
```

## Analyze Data:

```python
#explode user column
exploded = realArticlesPD.user.apply(pd.Series)
realArticlesPD = pd.concat([realArticlesPD.drop(columns='user'), exploded], axis=1)
#change name of entities to userEntities to avoid duplicate names
temp = realArticlesPD.columns.values; 
temp[36] = 'userEntities';
realArticlesPD.columns = temp
#explode userEntities
exploded = realArticlesPD.userEntities.apply(pd.Series)
realArticlesPD = pd.concat([realArticlesPD.drop(columns='userEntities'), exploded], axis=1)

#change name of URL and explode URL and description from entities
temp = realArticlesPD.columns.values; 
temp[64] = 'entitiesURL';
realArticlesPD.columns = temp
exploded = realArticlesPD['entitiesURL'][0].apply(pd.Series)
realArticlesPD = pd.concat([realArticlesPD.drop(columns='entitiesURL'), exploded], axis=1)
                            

#change name of description and explode entitiyDescription
#exploded = realArticlesPD.description.apply(pd.Series)
#realArticlesPD = pd.concat([realArticlesPD.drop(columns='description'), exploded], axis=1)
#add label to data
realArticlesPD['label'] = 'real'

#explode user column
exploded = fakeArticlesPD.user.apply(pd.Series)
fakeArticlesPD = pd.concat([fakeArticlesPD.drop(columns='user'), exploded], axis=1)
#change name of entities to userEntities to avoid duplicate names
temp = fakeArticlesPD.columns.values; 
temp[36] = 'userEntities';
fakeArticlesPD.columns = temp
#explode userEntities
exploded = fakeArticlesPD.userEntities.apply(pd.Series)
fakeArticlesPD = pd.concat([fakeArticlesPD.drop(columns='userEntities'), exploded], axis=1)


#change name of URL and explode URL and description from entities
temp = fakeArticlesPD.columns.values; 
temp[64] = 'entitiesURL';
fakeArticlesPD.columns = temp
exploded = fakeArticlesPD['entitiesURL'][0].apply(pd.Series)
fakeArticlesPD = pd.concat([fakeArticlesPD.drop(columns='entitiesURL'), exploded], axis=1)

#change name of description and explode entitiyDescription
#exploded = fakeArticlesPD.description.apply(pd.Series)
#fakeArticlesPD = pd.concat([fakeArticlesPD.drop(columns='description'), exploded], axis=1)
#add label to data
fakeArticlesPD['label'] = 'fake'

#combine PDs
bigdata = realArticlesPD.append(fakeArticlesPD, ignore_index=True)

#change name of id column to userID to avoid duplicate names
new_columns = bigdata.columns.values; 
new_columns[29] = 'userID'; 
new_columns[30] = 'userID_str'; 
bigdata.columns = new_columns

#drop zero non-null columns:
bigdata = bigdata.dropna(axis = 1, how = 'all')

#drop rows with less than 63 non-NAN values:
bigdata = bigdata.dropna(thresh=7) #t = 0 gives 1594 entries, t = 5 gives 1592 entries, t=10 gives 793 entries

#display head
bigdata.head()
```
