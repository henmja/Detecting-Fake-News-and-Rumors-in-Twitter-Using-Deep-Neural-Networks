## Read Data:

```python
import os, json
from os import path
import pandas as pd
import numpy as np
import glob
pd.set_option('display.max_columns', None)

fakeArticlesPD = pd.DataFrame()
realArticlesPD = pd.DataFrame()

for i in range(2):
    if i == 0:
        authenticity = 'fake'
    else:
        authenticity = 'real'
    rootdir = 'C:/Users/henri/FakenewsData/fakenewsnet_data/fakenewsnet_data/politifact/'+authenticity
    
    for subdir, dirs, articles in os.walk(rootdir):
        for article in dirs:
            temp = pd.DataFrame()
            path_to_json = 'C:/Users/henri/FakenewsData/fakenewsnet_data/fakenewsnet_data/politifact/' + authenticity + '/'+ article + '/tweets' 
            json_pattern = os.path.join(path_to_json,'*.json')
            file_list = glob.glob(json_pattern)

            for file in file_list:
                print(file)
                try:
                    data = pd.read_json(open(file, "r", encoding="utf8", errors="surrogateescape"), lines=True)
                    if authenticity == 'fake':
                        fakeArticlesPD = fakeArticlesPD.append(data, ignore_index = True)
                    else:
                        realArticlesPD = realArticlesPD.append(data, ignore_index = True)
                    break
                except ValueError:
                    print("ValueError")
        #print(temp["user"][0]['id'])
        #print(temp['retweets'][13][0]['user']['id'])
     
            path_to_json = 'C:/Users/henri/FakenewsData/fakenewsnet_data/fakenewsnet_data/politifact/'+authenticity+'/'+article+'/retweets' 
            json_pattern = os.path.join(path_to_json,'*.json')
            file_list = glob.glob(json_pattern)

            for file in file_list:
            #print(file)
                try:
                    data = pd.read_json(open(file, "r", encoding="utf8", errors="surrogateescape"), lines=True)
                    if authenticity == 'fake':
                        fakeArticlesPD = fakeArticlesPD.append(data, ignore_index = True)
                    else:
                        realArticlesPD = realArticlesPD.append(data, ignore_index = True)
                    break
                except ValueError:
                    print("ValueError")
```

## Analyze Data:

```python
#explode user column
exploded = realArticlesPD.user.apply(pd.Series)
realArticlesPD = pd.concat([realArticlesPD.drop(columns='user'), exploded], axis=1)
#change name of entities to userEntities to avoid duplicate names
temp = realArticlesPD.columns.values; 
temp[36] = 'userEntities';
realArticlesPD.columns = temp
#explode userEntities
exploded = realArticlesPD.userEntities.apply(pd.Series)
realArticlesPD = pd.concat([realArticlesPD.drop(columns='userEntities'), exploded], axis=1)

#print(realArticlesPD.head())

#change name of URL and explode URL and description from entities
temp = realArticlesPD.columns.values; 
temp[35] = 'tweetURL';
realArticlesPD.columns = temp
#print(realArticlesPD['entitiesURL'])
exploded = realArticlesPD.url.apply(pd.Series)
realArticlesPD = pd.concat([realArticlesPD.drop(columns='url'), exploded], axis=1)

#explode urls in column 0
exploded = realArticlesPD.urls.apply(pd.Series)
realArticlesPD = pd.concat([realArticlesPD.drop(columns='urls'), exploded], axis=1)
#change name of 0 to zero
temp = realArticlesPD.columns.values; 
temp[realArticlesPD.columns.get_loc(0)] = 'zero';
realArticlesPD.columns = temp

#add label to data
realArticlesPD['label'] = 'real'




#explode user column
exploded = fakeArticlesPD.user.apply(pd.Series)
fakeArticlesPD = pd.concat([fakeArticlesPD.drop(columns='user'), exploded], axis=1)
#change name of entities to userEntities to avoid duplicate names
temp = fakeArticlesPD.columns.values; 
temp[36] = 'userEntities';
fakeArticlesPD.columns = temp
#explode userEntities
exploded = fakeArticlesPD.userEntities.apply(pd.Series)
fakeArticlesPD = pd.concat([fakeArticlesPD.drop(columns='userEntities'), exploded], axis=1)


#change name of URL and explode URL and description from entities
temp = fakeArticlesPD.columns.values; 
temp[35] = 'tweetURL';
fakeArticlesPD.columns = temp
exploded = fakeArticlesPD.url.apply(pd.Series)
fakeArticlesPD = pd.concat([fakeArticlesPD.drop(columns='url'), exploded], axis=1)

#explode urls in column url
exploded = fakeArticlesPD.urls.apply(pd.Series)
fakeArticlesPD = pd.concat([fakeArticlesPD.drop(columns='urls'), exploded], axis=1)

#change name of 0 to zero
temp = fakeArticlesPD.columns.values; 
temp[fakeArticlesPD.columns.get_loc(0)] = 'zero';
fakeArticlesPD.columns = temp

#add label to data
fakeArticlesPD['label'] = 'fake'

#combine PDs
bigdata = realArticlesPD.append(fakeArticlesPD, ignore_index=True)

#change name of id column to userID to avoid duplicate names
new_columns = bigdata.columns.values; 
new_columns[29] = 'userID'; 
new_columns[30] = 'userID_str'; 
bigdata.columns = new_columns

#drop zero non-null columns:
bigdata = bigdata.dropna(axis = 1, how = 'all')

exploded = bigdata.zero.apply(pd.Series)
bigdata = pd.concat([bigdata.drop(columns='zero'), exploded], axis=1)
#fjern siste column
#bigdata = bigdata.drop(labels=0, axis=1)

#explode description
temp = bigdata.columns.values; 
temp[31] = 'tweetDescription';
bigdata.columns = temp
exploded = bigdata.description.apply(pd.Series)
bigdata = pd.concat([bigdata.drop(columns='description'), exploded], axis=1)

#explode quoted_status
exploded = bigdata.quoted_status.apply(pd.Series)
bigdata = pd.concat([bigdata.drop(columns='quoted_status'), exploded], axis=1)

#rename quoted_status items:
cols=pd.Series(bigdata.columns)
for dup in cols[cols.duplicated()].unique(): 
    cols[cols[cols == dup].index.values.tolist()] = [str(dup) + '_quoted_status' if i != 0 else str(dup) + '_tweet' for i in range(sum(cols == dup))]
# rename the columns with the cols list.
bigdata.columns=cols

#rename user
temp = bigdata.columns.values; 
temp[36] = 'created_at_entities_description'; #bigdata.columns.get_loc('created_at_quoted_status')
bigdata.columns = temp

#explode quoted_status_user column
exploded = bigdata.user.apply(pd.Series)
bigdata = pd.concat([bigdata.drop(columns='user'), exploded], axis=1)
#append name quoted_status_user
cols=pd.Series(bigdata.columns)
for dup in cols[cols.duplicated()].unique(): 
    cols[cols[cols == dup].index.values.tolist()] = [str(dup) + '_quoted_status_user' if i != 0 else dup for i in range(sum(cols == dup))]
# rename the columns with the cols list.
bigdata.columns=cols

#explode entities_tweet
#explode entities column
exploded = bigdata.entities_tweet.apply(pd.Series)
bigdata = pd.concat([bigdata.drop(columns='entities_tweet'), exploded], axis=1)
#append name entities
cols=pd.Series(bigdata.columns)
for dup in cols[cols.duplicated()].unique(): 
    cols[cols[cols == dup].index.values.tolist()] = [str(dup) + '_entities' if i != 0 else dup for i in range(sum(cols == dup))]
# rename the columns with the cols list.
bigdata.columns=cols

#explode entities_quoted_status
exploded = bigdata.entities_quoted_status.apply(pd.Series)
bigdata = pd.concat([bigdata.drop(columns='entities_quoted_status'), exploded], axis=1)
#append name entities_quoted_status
cols=pd.Series(bigdata.columns)
for dup in cols[cols.duplicated()].unique(): 
    cols[cols[cols == dup].index.values.tolist()] = [str(dup) + '_entities_quoted_status' if i != 0 else dup for i in range(sum(cols == dup))]
# rename the columns with the cols list.
bigdata.columns=cols

#explode entities_user_quoted_status
exploded = bigdata.entities.apply(pd.Series)
bigdata = pd.concat([bigdata.drop(columns='entities'), exploded], axis=1)
#append name entities_quoted_status
cols=pd.Series(bigdata.columns)
for dup in cols[cols.duplicated()].unique(): 
    cols[cols[cols == dup].index.values.tolist()] = [str(dup) + '_quoted_status' if i != 0 else dup for i in range(sum(cols == dup))]
# rename the columns with the cols list.
bigdata.columns=cols

#explode hashtags
exploded = bigdata.hashtags.apply(pd.Series)
bigdata = pd.concat([bigdata.drop(columns='hashtags'), exploded], axis=1)
#append name entities_quoted_status
cols=pd.Series(bigdata.columns)
for dup in cols[cols.duplicated()].unique(): 
    cols[cols[cols == dup].index.values.tolist()] = [str(dup) + '_hashtags_entities_quoted_status' if i != 0 else dup for i in range(sum(cols == dup))]
# rename the columns with the cols list.
bigdata.columns=cols


#drop zero non-null columns:
bigdata = bigdata.dropna(axis = 1, how = 'all')

#drop rows with less than 63 non-NAN values:
bigdata = bigdata.dropna(thresh=7) #t = 0 gives 1594 entries, t = 5 gives 1592 entries, t=10 gives 793 entries

#movel label to end:
#cols = list(df.columns.values) #Make a list of all of the columns in the df
#cols.pop(cols.index('b')) #Remove b from list
#cols.pop(cols.index('x')) #Remove x from list
#df = df[cols+['b','x']]

#display head
bigdata.head()
```
